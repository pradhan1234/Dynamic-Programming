```python
# Naive recursive algorithm to compute fib of n, where f(n) = f(n-1) + f(n-2), f(0) = 0, and f(1) = 1.
import datetime

def fib(n):
    if(n == 0 or n == 1):
        return n
    else:
#         print(n)
#         print(factorial(n-1) + factorial(n-2))
        return (fib(n-1) + fib(n-2))



#using dynamic programming
def dynamicFib(n):
    A = []
    A.append(0)
    A.append(1)
    for i in range(2,n+1):
        A.append(A[i-1] + A[i-2])
    return A[n]
```

```python
num = 10000
```

```python
t1 = datetime.datetime.now()
print(fib(num))
t2 = datetime.datetime.now()
print(t2-t1)
```

```python
t1 = datetime.datetime.now()
print(dynamicFib(num))
t2 = datetime.datetime.now()
print(t2-t1)
```
